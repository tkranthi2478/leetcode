```python
class Trie:

    def __init__(self):

        self.root = {}

    def insert(self, word):

        cur = self.root
        for letter in word:
            if letter not in cur:
                cur[letter] = {}
            cur = cur[letter]
        cur['*'] = ''

    def search(self, word):
        cur = self.root
        for letter in word:
            if letter not in cur:
                return False
            cur = cur[letter]
        return '*' in cur
        
    def startsWith(self, prefix):
        cur = self.root
        for letter in prefix:
            if letter not in cur:
                return False
            cur = cur[letter]
        return True

    def w_search(self, word):
        # assuming * stands for any char -- only one occurence
        return self.rec_search(word, self.root)

    def rec_search(self, word, cur): # internal
        if word:
            letter = word[0]
            if letter == '*':
                # exhaust all possibilities of start's children
                return any(self.rec_search(word[1:], cur[child]) for child in cur)
            elif letter not in cur:
                return False
            else:
                return self.rec_search(word[1:], cur[letter])
        else:
            return '#' in cur
    
    def search2(self, word):
        return self.rec_search2(word, self.root)

    def rec_search2(self, word, cur): # internal
        # assuming * stands for any char -- any number of occurence
        if word:
            letter = word[0]
            if letter == '*':
                # exhaust all possibilities of start's children
                return any(self.rec_search2(word[1:], cur[child]) for child in cur) or any(self.rec_search2(word, cur[child]) for child in cur)
            elif letter not in cur:
                return False
            else:
                return self.rec_search2(word[1:], cur[letter])
        else:
            return '#' in cur

trte = Trie()
words = ["apple", "apped", "appled"]
for word in words:
    trte.insert(word)
print(trte.search("app*e"))
print(trte.search("appl*"))
print(trte.search("app*d"))
print(trte.search("app**"))
print(trte.search("app**k"))
print(trte.search("********"))
print(trte.search2("a*e"))
print(trte.search2("*"))
print(trte.search2("a*l*"))
```

```python
# https://leetcode.com/problems/find-the-length-of-the-longest-common-prefix/description/
def longestCommonPrefix(arr1, arr2):
        root = {}
        for num in arr1:
            cur = root
            for char in str(num):
                if char not in cur:
                    cur[char] = {}
                cur = cur[char]
        max_len = 0
        for num in arr2:
            cur = root
            i = 0
            for char in str(num):
                if char in cur:
                    cur = cur[char]
                    i+=1
                else:
                    break
            max_len = max(max_len, i)
        return max_len
```

```python
# https://leetcode.com/problems/design-add-and-search-words-data-structure/description/

class WordDictionary:

    def __init__(self):
        self.root = {}
        
    def addWord(self, word: str) -> None:
        cur = self.root
        for char in word:
            if char not in cur:
                cur[char] = {}
            cur = cur[char]
        cur['*']= {}

    def search(self, word: str) -> bool: 
        def rec_search(cur, word, i): 
            if i==len(word):
                return '*' in cur
            elif word[i] == '.':
                return any([rec_search(val, word, i+1) for val in cur.values()])
            elif word[i] not in cur:
                return False
            else:
                return rec_search(cur[word[i]], word, i+1)
        return rec_search(self.root, word, 0)


# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```
