```python
# ğŸ’¬ â€œEach node in a linear network must run the same program and discover the total number of nodes using asynchronous message passing. How would you implement it?â€

import time

class Message:
    def __init__(self, count):
        self.count = count

class NodeCount:
    def __init__(self):
        self.count = -1

    def is_left_most(self):
        return False

    def is_right_most(self):
        return False

    def send_left(self, message):
        pass

    def send_right(self, message):
        pass

    def get_count(self):
        if self.is_left_most():
            self.send_right(Message(1))
        while self.count == -1:
            time.sleep(0.1)
        return self.count

    def receive_from_left(self, message):
        if self.is_right_most():
            message.count += 1
            self.send_left(message)
        else:
            self.send_right(Message(message.count + 1))

    def receive_from_right(self, message):
        if self.is_left_most():
            self.count = message.count
        else:
            self.send_left(message)
```
