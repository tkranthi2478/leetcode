## Rules of Binary Search(L BISECT)
1. Left, Right are included
2. (l+r)//2 -- gives left biased middle element
3.          
                  func(m) < target    STRICT EQUALITY
                      /       \
                  NO           YES
                  /               \
            [l, m]             [m+1, r]
      (i.e. l = m+1)      (i.e. l = m+1)

````python

from bisect import bisect_left, bisect_right

def l_bisect(nums, target: int) -> int:
    if not nums:
        return 0
    l, r = 0, len(nums)
    while l < r:
        m = (l + r) // 2
        if nums[m] < target:
            l = m + 1
        else:
            r = m
    return l

assert(l_bisect([1,2,5,5,5,5,5,6,7,8,9,9,9,9,9,10], 5) == bisect_left([1,2,5,5,5,5,5,6,7,8,9,9,9,9,9,10], 5)==2)
assert(l_bisect([], 1) == bisect_left([], 1) ==0)
assert(l_bisect([1], 2) == bisect_left([1], 2) == 1 )
assert(l_bisect([1,2,2], 2)  == bisect_left([1], 2) ==1)
assert(l_bisect([1,2], 3) == bisect_left([1,2], 3)==2)
assert(l_bisect([1,2], -3) == bisect_left([1,2], -3)==0)
assert(l_bisect([1,3], 2) == bisect_left([1,3], 2)==1)


def r_bisect( nums, target: int) -> int:  
    if not nums:
        return 0
    l, r = 0, len(nums)
    while l < r:
        m = (l + r) // 2 
        if nums[m] <= target:
            l = m+1
        else:
            r = m
    return l


assert(r_bisect([1,2,5,5,6,7,8,9,9,9,9,9,10], 5) == bisect_right([1,2,5,5,6,7,8,9,9,9,9,9,10], 5)==4)
assert(r_bisect([], 1) == bisect_right([], 1)== 0)
assert(r_bisect([1], 2) == bisect_right([1], 2) == 1)
assert(r_bisect([1,2,2], 2) == bisect_right([1,2,2], 2)== 3)
assert(r_bisect([1,2], 3) == bisect_right([1,2], 3)== 2)
assert(r_bisect([1,2], -3) == bisect_right([1,2], -3) == 0)
assert(r_bisect([1,3], 2) == bisect_right([1,3], 2) == 1)
# nums = [1,2]
````

Binary Section -- Thinking
1. Monotonic Function -- Tipping Point -- k such that f(k) < OPT and f(k+1)>OPT
2. Interval will always contain the optimal -- 1 size means l == r
3. Every Bisect problem we have seen l_bisect
4. Boolean Critiera
       1. Obvious case f(m) < l
       2. Non Obvious case -- see if soln is to right of mid

````python
# Binary Search -- Exact match
def search(nums, target) -> int:
    l, r = 0, len(nums) - 1
    while l < r:
        m = (l + r) // 2
        if nums[m] < target:
            l = m + 1
        else:
            r = m
    return l if l < len(nums) and nums[l] == target else -1

# https://leetcode.com/problems/cutting-ribbons/
def maxLength(ribbons, k):
    total = sum(ribbons)
    if total < k:
        return 0
    l, r = 1, total // k
    while l < r:
        m = (l + r ) // 2  # upper mid to prevent infinite loop
        if sum(v // m for v in ribbons) <= k:
            r = m 
        else:
            l = m + 1
    return l

````

````python
#  https://leetcode.com/problems/find-peak-element/
#  Why is this not in the exact section?
#  We are effectively trying to find 2 elt range with l > l+1


def findAnyPeakElement(nums) -> int:
    l, r = 0, len(nums)-1
    while l<r:
        mid = (l+r)//2
        if nums[mid]<nums[mid+1]:
            l = mid+1
        else:
            r = mid
    return l

#Fnding the closest elmenets
def findkClosestElements(arr, k, x):
    l,r = 0,len(arr)-k
    while l < r:
        mid = (l+r)//2
        #we are evaluating if window [mid, mid+k-1] is better than [mid+1, mid+k]
        #by comparing l's distance with mid and mid+k
        if x-arr[mid] > arr[mid+k]-x:
            l = mid + 1
        else:
            r = mid  
    return arr[l:l+k]

def closetElt( nums, target) -> int:
    l,r  = 0, len(nums)-1
    while l < r:
        mid = (l+r)//2
        #we are evaluating if window [mid] is better than [mid+1]
        #by comparing l's distance with m
        if target-nums[mid] > nums[mid+1]-target:
            l = mid + 1
        else:
            r = mid  
    return nums[l:l+1]
````

````python
# binary search 
# min sum(boats) and m

def shipWithinDays(weights, D):
    # Utils function to check if a given capacity can ship within D days
    def can_ship(capacity):
        days, current_load = 1, 0
        for w in weights: # It palces the weights in the boat in the order they are given until load exceeds capacity
            if current_load + w > capacity:
                days += 1
                current_load = 0
            current_load += w
        return days <= D


    left, right = max(weights), sum(weights)    
    # binary search -- is efficent at find the optimal value
    # with binary search we shrink the available soln space 
    # initial it is [l, r] -- full range
    # if m satisfies the condition  the next bounds are [left, m]
    # else the soln space is [m+1, right]
    # we shrink the available space until it is left == right

    # intiially [ l,r]  is [ 4, 17]
    # after 1st iteration[ 4,16]'s  mid = 10 satisfies thus  [ 4,16] ---becomes---> [ 4, 10]
    # after 2nd iteration [ 4,10]'s mid = 7 satisfies this [ 4,10] --> [ 4, 7]
    # after 3rd iteration [ 4,7]'s  mid = 7 doesnt ship this [ 4,7] --> [ 6, 7]
    # after 4th iteration [ 6,7]'s  mid = 6 doesnt ship this [ 6,7] --> [ 7, 7]
    # return 7

    while left < right:
        print(left, right)
        mid = (left + right) // 2
        if can_ship(mid):  # mid can sastisfy the condition
            print("mid:", mid, "can ship")
            right = mid  # yes we [ left, right] --> [ left, mid]
        else:
            print("mid:", mid, "cant ship")
            left = mid + 1 # [ left, right] --> [mid+1, right]
    return left
print(shipWithinDays([3,2,2,1,4,1,4], 3)) # 15
````
