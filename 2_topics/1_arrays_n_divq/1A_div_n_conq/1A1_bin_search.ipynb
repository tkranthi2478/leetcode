````python
# Binary Search -- Exact match
def search( nums, target) -> int:
    left, right = 0, len(nums) - 1
    while left <= right:
        m = (left + right) // 2
        if nums[m] == target:
            return m
        if nums[m] < target:
            left = m + 1
        else:
            right = m - 1
    return -1

# https://leetcode.com/problems/cutting-ribbons/
def maxLength(ribbons, k):
    total = sum(ribbons)
    if total < k:
        return 0
    l, r = 1, total // k
    while l <= r:
        m = (l + r) // 2
        if sum([v // m for v in ribbons]) < k:
            r = m - 1
        else:
            l = m + 1
    return r

````

````python

from bisect import bisect_left, bisect_right

def l_bisect(nums, target: int) -> int:
    if not nums:
        return 0
    l, r = 0, len(nums)
    while l < r:
        m = (l + r) // 2
        if nums[m] < target:
            l = m + 1
        else:
            r = m
    return l

assert(l_bisect([1,2,5,5,5,5,5,6,7,8,9,9,9,9,9,10], 5) == bisect_left([1,2,5,5,5,5,5,6,7,8,9,9,9,9,9,10], 5)==2)
assert(l_bisect([], 1) == bisect_left([], 1) ==0)
assert(l_bisect([1], 2) == bisect_left([1], 2) == 1 )
assert(l_bisect([1,2,2], 2)  == bisect_left([1], 2) ==1)
assert(l_bisect([1,2], 3) == bisect_left([1,2], 3)==2)
assert(l_bisect([1,2], -3) == bisect_left([1,2], -3)==0)
assert(l_bisect([1,3], 2) == bisect_left([1,3], 2)==1)


def r_bisect( nums, target: int) -> int:  
    if not nums:
        return 0
    l, r = 0, len(nums)
    while l < r:
        m = (l + r) // 2 
        if nums[m] <= target:
            l = m+1
        else:
            r = m
    return l


assert(r_bisect([1,2,5,5,6,7,8,9,9,9,9,9,10], 5) == bisect_right([1,2,5,5,6,7,8,9,9,9,9,9,10], 5)==4)
assert(r_bisect([], 1) == bisect_right([], 1)== 0)
assert(r_bisect([1], 2) == bisect_right([1], 2) == 1)
assert(r_bisect([1,2,2], 2) == bisect_right([1,2,2], 2)== 3)
assert(r_bisect([1,2], 3) == bisect_right([1,2], 3)== 2)
assert(r_bisect([1,2], -3) == bisect_right([1,2], -3) == 0)
assert(r_bisect([1,3], 2) == bisect_right([1,3], 2) == 1)
# nums = [1,2]
````

````python
#  https://leetcode.com/problems/find-peak-element/
#  Why is this not in the exact section?
#  We are effectively trying to find 2 elt range with l > l+1


def findAnyPeakElement(nums) -> int:
    l, r = 0, len(nums)-1
    while l<r:
        mid = (l+r)//2
        if nums[mid]<=nums[mid+1]:
            l = mid+1
        else:
            r = mid
    return l

#Fnding the closest elmenets
def findkClosestElements(arr, k, x):
    l,r = 0,len(arr)-k
    while l < r:
        mid = (l+r)//2
        #we are evaluating if window [mid, mid+k-1] is better than [mid+1, mid+k]
        #by comparing l's distance with mid and mid+k
        if x-arr[mid] > arr[mid+k]-x:
            l = mid + 1
        else:
            r = mid  
    return arr[l:l+k]

def closetElt( nums, target) -> int:
    l,r  = 0, len(nums)-1
    while l < r:
        mid = (l+r)//2
        #we are evaluating if window [mid] is better than [mid+1]
        #by comparing l's distance with m
        if target-nums[mid] > nums[mid+1]-target:
            l = mid + 1
        else:
            r = mid  
    return nums[l:l+1]
````
